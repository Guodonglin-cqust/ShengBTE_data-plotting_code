```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1=np.loadtxt('BTE.WP3')
print(f1.size)
print(f1.shape)
#print(f1)
```

```python
plt.plot(f1[:,0],f1[:,1],"x")
plt.show()
```

```python
f2=np.loadtxt('BTE.WP3_plus')
print(f2.size)
print(f2.shape)
#print(f2)
```

```python
plt.plot(f2[:,0],f2[:,1],"x")
plt.show()
```

```python
f3=np.loadtxt('BTE.WP3_minus')
print(f3.size)
print(f3.shape)
#print(f3)
```

```python
plt.plot(f3[:,0],f3[:,1],"x")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(f1[:,0],f1[:,1],"x",label="P$_3$(Total)")
plt.plot(f2[:,0],f2[:,1],"x",label="P$_3$$^+$(Absorption)")
plt.plot(f3[:,0],f3[:,1],"x",label="P$_3$$^-$(Emission)")
plt.legend(loc=1,fontsize=12)
plt.xlabel('Frequency (rad/ps)',size=12)
plt.ylabel('Weighted phase space(ps$^4$/rad$^4$)',size=12)
plt.show()
#plt.savefig('Scattering_ratevsFrequency-1.png',dpi=500)
```

```python
plt.figure(figsize=(8,6))
plt.plot(f1[:,0]/6.28,f1[:,1],"x",label="P$_3$(Total)")
plt.plot(f2[:,0]/6.28,f2[:,1],"x",label="P$_3$$^+$(Absorption)")
plt.plot(f3[:,0]/6.28,f3[:,1],"x",label="P$_3$$^-$(Emission)")
plt.legend(loc=1,fontsize=12)
plt.xlabel('Frequency (THz)',size=12)
plt.ylabel('Weighted phase space(ps$^4$/rad$^4$)',size=12)
#plt.show()
plt.savefig('Weighted_phase_spacevsFrequency-1.png',dpi=500)
```
